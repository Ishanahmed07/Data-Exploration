```python
sl = data.columns[data.sum()<5_000]
print(data[sl].head())

plt.figure(figsize=(8,8))
x_pos = np.arange(len(data[sl].columns))
plt.bar(x_pos,data[sl].sum(),align='center')
plt.xticks(x_pos,data[sl].columns)
plt.show()
```

```python
categories  = ['High','Medium','Low']
categories_selected = [[] for i in range(len(categories))]
for name in data.columns:
  total_hits = data[name].sum()
  if total_hits > 100_000:
    category = 0
  elif total_hits > 5_000:
    category = 1
  else:
    category = 2
  categories_selected[category].append(name)
  print(f'Page {name} is {categories[category]} volume')

for i in range(len(categories)):
  print(f'{categories[i]} : {str(categories_selected[i])}')

for i, selected in enumerate(categories_selected):
  plt.figure(figsize=(8,8))
  x_pos = np.arange(len(data[selected].columns))
  plt.bar(x_pos,data[selected].sum(),align='center')
  plt.xticks(x_pos,data[selected].columns)
  plt.xlabel('Pages',fontsize=18)
  plt.ylabel('Hits',fontsize=18)
  plt.title(f'{categories[i]} Volume Pages',fontsize=20)
  plt.show()
```

```python
explodeList = []
selecte = []
columns = data.columns
data['Others'] = [0] * len(data.index)
for name in columns:
  total_hits = data[name].sum()
  if total_hits > 5_000:
    selecte.append(name)
    explodeList.append(0)
  else:
    data['Others'] += data[name]

selecte.append('Others')
explodeList.append(0.05)
print(data[selecte].head())

plt.figure(figsize=(8,8))
plt.pie(data[selecte].sum(), labels=selecte,explode=explodeList,autopct='%1.1f%%',startangle = 90)
plt.title('Total Page Hits', fontsize = 20)
plt.legend(loc = 2)
plt.show()
```
